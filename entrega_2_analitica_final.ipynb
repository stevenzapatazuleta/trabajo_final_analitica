```python
duracion_desem=pd.read_csv('/content/drive/MyDrive/Entregable 2 Analítica/Duration of unemployment.csv', sep=',')
duracion_desem
duracion_desem['MONTH'] = duracion_desem['MONTH'].astype('string')
duracion_desem['MONTH'] = duracion_desem['MONTH'].apply(lambda x: x[0:3] +'/'+x[3:7])

#Dividir el dataset con el delimitador "espacio"
separado = duracion_desem['MONTH'].str.split("/", n=1, expand=True)

#Al resultado del split lo ponemos en la primera columna
duracion_desem['MONTH']= separado[0]

#Hacemos lo mismo con la segunda columna
duracion_desem['fecha']= separado[1]

#Conversión a int
duracion_desem['fecha']=duracion_desem['fecha'].astype('int')

#Conversión tipo de dato de object a string para manipulación de datos
duracion_desem.loc[:, ('GEOGRAPHY', 'DURATION','AGE GROUP','MONTH')]=duracion_desem.loc[:, ('GEOGRAPHY', 'DURATION','AGE GROUP','MONTH')].astype('string')

#String en minúscula
duracion_desem['GEOGRAPHY']=duracion_desem ['GEOGRAPHY'].str.lower()
duracion_desem['DURATION']=duracion_desem ['DURATION'].str.lower()
duracion_desem['AGE GROUP']=duracion_desem ['AGE GROUP'].str.lower()
duracion_desem['MONTH']=duracion_desem ['MONTH'].str.lower()

#Reondear y convertir a entero
duracion_desem['Female']=duracion_desem ['Female'].round()
duracion_desem['Male']=duracion_desem ['Male'].round()
duracion_desem['Both sexes']=duracion_desem ['Both sexes'].round()
duracion_desem.loc[:, ('Female','Male','Both sexes')]=duracion_desem.loc[:, ('Female','Male','Both sexes')].astype(int)

#Renombrar columnas
duracion_desem = duracion_desem.rename(columns={'MONTH':'mes','GEOGRAPHY':'geografia','DURATION':'duracion','AGE GROUP':'rango edad','Both sexes':'total personas','Male':'hombres','Female':'mujeres'})

#Acotar rango de fecha
duracion_desem=duracion_desem[duracion_desem['fecha']<=2015]

#Renombrar datos por facilidad en visualización
duracion_desem['duracion']=duracion_desem['duracion'].replace('average weeks unemployed (no top-code)','average unemployed nc').replace('average weeks unemployed (top-code = 99 weeks)','average unemployed c99')
duracion_desem
```

## **Nivel educativo**

```python
rates= pd.read_csv('/content/drive/MyDrive/Entregable 2 Analítica/Unemployment rates of 25- to 29-year-olds, by educational attainment, Canada and provinces.csv', sep=',')
rates1 = rates.copy()
rates1.rename(columns={'REF_DATE':'fecha', 'Educational attainment': 'nivel_educativo','GEO':'geografia', 'VALUE':'valor'},inplace=True)
rates1['geografia']=rates1['geografia'].str.lower()
rates1['nivel_educativo']=rates1['nivel_educativo'].str.lower()
rates1['nivel_educativo']=rates1['nivel_educativo'].astype('string')
rates1['nivel_educativo']=rates1['nivel_educativo'].astype('string')
rates1['valor']=rates1['valor'].fillna(0)
#relleno prom
rates_education=rates1.drop(['DGUID', 'UOM', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS'], axis=1).reset_index().drop('index', axis=1)
rates_education = rates_education[(rates_education['fecha']<=2015) & (rates_education['fecha']>=2001)]
rates_education1 = rates_education[rates_education['nivel_educativo']== 'less than high school']
rates_education2 = rates_education[rates_education['nivel_educativo']== 'all levels of education']
rates_education3 = rates_education[rates_education['nivel_educativo']== 'high school']
rates_education4 = rates_education[rates_education['nivel_educativo']== 'college or trade']
rates_education5 = rates_education[rates_education['nivel_educativo']== 'university']
def funcion(x):
  if x['valor'] == 0 and x['nivel_educativo']== 'less than high school' :
    valor =rates_education1['valor'].mean()
  elif x['valor'] == 0 and x['nivel_educativo'] =='all levels of education':
    valor =rates_education2['valor'].mean()
  elif x['valor'] == 0 and x['nivel_educativo'] =='high school':
    valor =rates_education3['valor'].mean()
  elif x['valor'] == 0 and x['nivel_educativo'] =='college or trade':
    valor =rates_education4['valor'].mean()
  elif x['valor'] == 0 and x['nivel_educativo'] =='university':
    valor = rates_education5['valor'].mean()
  else:
    valor = x['valor']
  return valor

rates_education['valor'] =rates_education.apply(funcion, axis=1)
rates_education
```

## **Pregunta 1**

¿Cuál región de Ontario presenta el peor indicador de duración del desempleo anualmente?¿Cambia si se realiza por trimestre? (La forma de sacar el indicador es establecido previamente)

***Dataframe***

```python
duracion_desem_p1=duracion_desem.copy()

#Eliminando categorías que no son de interes para esta pregunta
duracion_desem_p1=duracion_desem_p1[duracion_desem_p1['duracion']!='total unemployed']
duracion_desem_p1=duracion_desem_p1[duracion_desem_p1['duracion']!='   duration unknown']
duracion_desem_p1=duracion_desem_p1[duracion_desem_p1['duracion']!='average unemployed nc']
duracion_desem_p1=duracion_desem_p1[duracion_desem_p1['duracion']!='average unemployed c99']
duracion_desem_p1=duracion_desem_p1[duracion_desem_p1['geografia']!='total, ontario regions']
duracion_desem_p1['duracion']=duracion_desem_p1['duracion'].replace('   27 weeks or more','      27 - 51 weeks')
duracion_desem_p1=duracion_desem_p1.rename(columns={'mes':'trimestre'})
duracion_desem_p1=duracion_desem_p1[(duracion_desem_p1['rango edad']!='total, 15 years and over')&(duracion_desem_p1['rango edad']!=' 15-64 years')]
duracion_desem_p1
```

```python
#Creando indicador, dependiendo de la categoría de la duranción se dara un valor de criticidad y posteriormente se multiplica dicho valor por la cantidad de personas
def funcion(x):
  if x['duracion'] =='   1 - 4 weeks':
    valor= 1
  elif x['duracion'] =='   5 - 13 weeks':
    valor=2
  elif x['duracion'] =='   14 - 25 weeks':
    valor=3   
  elif x['duracion'] =='   26 weeks ':
    valor=4
  elif x['duracion'] =='      27 - 51 weeks':
    valor=5
  elif x['duracion'] =='      52 weeks':
    valor=6 
  else:
    valor=7
  return valor 

duracion_desem_p1['criticidad'] = duracion_desem_p1.apply(funcion, axis = 1)
duracion_desem_p1['indicador']=duracion_desem_p1['total personas']*duracion_desem_p1['criticidad']
duracion_desem_p1['indicador']=duracion_desem_p1['indicador']/1000

#Dataframe con indicador de duración de desempleo por región y año
duracion_desem_p11 = duracion_desem_p1.groupby(['geografia','fecha'])[['indicador']].sum()
duracion_desem_p11=duracion_desem_p11.reset_index()
duracion_desem_p11
```

```python
#Dataframe con la región con el peor indicador de duración de desempleo por año 
duracion_desem_p111=duracion_desem_p11.copy()
duracion_desem_p111=duracion_desem_p111.sort_values('indicador',ascending=False)
duracion_desem_p111=duracion_desem_p111.drop_duplicates(['fecha'],keep='first')
duracion_desem_p111=duracion_desem_p111.sort_values('fecha', ascending=True)
duracion_desem_p111
```

```python
#Dataframe con indicadores por trimestre
duracion_desem_p12= duracion_desem_p1.groupby(['geografia','trimestre'])[['indicador']].sum().reset_index()
duracion_desem_p12['trimestre']=duracion_desem_p12['trimestre'].replace('dec',4).replace('sep',3).replace('jun',2).replace('mar',1)
duracion_desem_p12
```

```python
#Dataframe con la región con el peor indicador de duración de desempleo por trimestre 
duracion_desem_p122=duracion_desem_p12.copy()
duracion_desem_p122=duracion_desem_p122.sort_values('indicador',ascending=False)
duracion_desem_p122=duracion_desem_p122.drop_duplicates(['trimestre'],keep='first')
duracion_desem_p122=duracion_desem_p122.sort_values('trimestre', ascending=True)
duracion_desem_p122
```

## **Pregunta 2**

¿Cuál es la correlación del desempleo por tipo de nivel educativo y el pib anual?

```python
#Dataframe porcentaje de desempleor por Nivel Educativo
rates_education_p2=rates_education[rates_education['nivel_educativo']!='all levels of education']
rates_education_p2=rates_education_p2[rates_education_p2['geografia']=='ontario'].reset_index()
rates_education_p2=rates_education_p2.drop(['index'],axis=1)
rates_education_p2
```

```python
#Dataframe PIB
pib_def_p2=pib_def[['fecha','unidad de medida', 'valor']]
pib_def_p2=pib_def_p2.rename(columns={'valor':'monto'})
pib_def_p2=pib_def_p2.reset_index()
pib_def_p2=pib_def_p2.drop(['index'],axis=1)
pib_def_p2
```

```python
#Dataframe concatenado
pp2=pd.merge(rates_education_p2, pib_def_p2, how = 'left', on = 'fecha')
pp2
```

```python
array1=[]
cor1=pp2[pp2['nivel_educativo']=='less than high school']
cor2=pp2[pp2['nivel_educativo']=='high school']
cor3=pp2[pp2['nivel_educativo']=='college or trade']
cor4=pp2[pp2['nivel_educativo']=='university']
array1.append([cor1['monto'].corr(cor1['valor'])])
array1.append([cor2['monto'].corr(cor2['valor'])])
array1.append([cor3['monto'].corr(cor3['valor'])])
array1.append([cor4['monto'].corr(cor4['valor'])])
array1
```

```python
def funcion(x):
  if x['nivel_educativo'] =='less than high school':
    valor = array1[0]
  elif  x['nivel_educativo'] =='high school':
    valor = array1[1]
  elif  x['nivel_educativo'] =='college or trade':
    valor = array1[2]
  else:
    valor = array1[3]
  return valor 
pp2['correlacion'] = pp2.apply(funcion, axis = 1)
pp2['correlacion'] =pp2['correlacion'].astype('string')
pp2['correlacion']=pp2['correlacion'].str.replace("[","").str.replace("]","")
pp2['correlacion']=pp2['correlacion'].apply(lambda x: x[0:5])
pp2
```

***Gráfico***

```python
#Transformación sobre el score para visualizar mayor diferencia entre el tamaño de los puntos
pp2['valor_norm'] = (pp2['valor'] - pp2['valor'].min())/(pp2['valor'].max()- pp2['valor'].min())

#Gráfica
fig = px.scatter(pp2, x = 'valor', y ='monto', color = 'nivel_educativo', title = '<b>Desempleo por nivel educativo vs PIB anual<b>',
           hover_name = 'nivel_educativo', size = 'valor_norm', hover_data = ['correlacion'])

#Atributos
fig.update_layout(
    xaxis_title = '<b>Desempleo por nivel educativo<b>',
    yaxis_title = '<b>PIB anual<b>',
    template = 'simple_white',
    title_x = 0.5,)
fig.show()
```

```python
# plotting the figure
fig = px.scatter_3d(pp2, x="correlacion", y="valor", z="monto", color = 'nivel_educativo', title = '<b>Desempleo por nivel educativo vs PIB anual<b>',
           hover_name = 'nivel_educativo', size = 'valor_norm')
fig
```

***Interpretación***
Entre el porcentaje de desempleo por nivel educativo y el PIB anual existe una correlación positiva baja en la mayoría de categorías, como lo son personas que no han terminado sus estudios escolares, los que han terminado el colegio y personas que entran a escuelas de comercio. Por otra parte no se logra apreciar una correlación entre el PIB y las personas que tienen estudios universitarios.

También se evidencia que entre mayor sea el nivel de educación de la persona el porcentaje de desempleo es más bajo, lo que indica una mayor oportunidad laboral, sin embargo los valores entre universitarios y personas que tienen estudios de escuelas de comercio es muy similar, particularidad que en artículos se debate que en la actualidad las personas han empezado a optar por alternativas a los títulos universitarios y que pueden brindarles un beneficio similar o igual.

Otra comportamiento evidenciado son los rangos del porcentaje de desempleo por nivel educativo, que son más amplios para estudios inferiores. De acuerdo a la correlación obtenida se infiere que la tecnología usada en los diversos sectores industriales ha aumentado el PIB sin necesidad de que el recurso humano crezca en la misma medida, al punto de prescindir de personas que no cuenten con los estudios necesarios, planteamiento que puede explicar porque a medida que el PIB aumenta el porcentaje de desempleados en los niveles educativos inferiores crece.

## **Pregunta 7**
¿Existen patrones o comportamientos especiales en la desempleabilidad por trimesre y por año con base en los grupos por edades?

```python
duracion_desem

# Ignorar categoria que representa el totalizado de las otras categorias
duracion_desem = duracion_desem[duracion_desem['duracion']!='   27 weeks or more'] 
duracion_desem7 = duracion_desem[(duracion_desem['rango edad']!= 'total, 15 years and over') & (duracion_desem['rango edad']!= ' 15-64 years')] 
duracion_desem7 

# bases con el total de desempleo por trimestre en cada rango de edad 
desem_rango1 = duracion_desem7[duracion_desem7['rango edad']=='  15-19'].groupby(['mes'])[['total personas']].sum().reset_index()
desem_rango2 = duracion_desem7[duracion_desem7['rango edad']=='  20-24'].groupby(['mes'])[['total personas']].sum().reset_index()
desem_rango3 = duracion_desem7[duracion_desem7['rango edad']=='  25-44'].groupby(['mes'])[['total personas']].sum().reset_index()
desem_rango4 = duracion_desem7[duracion_desem7['rango edad']=='  45-54'].groupby(['mes'])[['total personas']].sum().reset_index()
desem_rango5 = duracion_desem7[duracion_desem7['rango edad']=='  55-64'].groupby(['mes'])[['total personas']].sum().reset_index()
desem_rango6 = duracion_desem7[duracion_desem7['rango edad']=='  65 years and over'].groupby(['mes'])[['total personas']].sum().reset_index()

desem_rangos = pd.merge(desem_rango1, desem_rango2, how = 'left', on = 'mes').merge(desem_rango3, how = 'left', on = 'mes').merge(desem_rango4, how = 'left', on = 'mes').merge(desem_rango5, how = 'left', on = 'mes').merge(desem_rango6, how = 'left', on = 'mes')
desem_rangos.columns = ['trimestre','15-19','20-24','25-44','45-54','55-64','65 years and over']
desem_rangos # desempleo por trimestre segun el rango de edad 
```

```python
# bases con el total de desempleo por año en cada rango de edad 
desem_rango1a = duracion_desem7[duracion_desem7['rango edad']=='  15-19'].groupby(['fecha'])[['total personas']].sum().reset_index()
desem_rango2a = duracion_desem7[duracion_desem7['rango edad']=='  20-24'].groupby(['fecha'])[['total personas']].sum().reset_index()
desem_rango3a = duracion_desem7[duracion_desem7['rango edad']=='  25-44'].groupby(['fecha'])[['total personas']].sum().reset_index()
desem_rango4a = duracion_desem7[duracion_desem7['rango edad']=='  45-54'].groupby(['fecha'])[['total personas']].sum().reset_index()
desem_rango5a = duracion_desem7[duracion_desem7['rango edad']=='  55-64'].groupby(['fecha'])[['total personas']].sum().reset_index()
desem_rango6a = duracion_desem7[duracion_desem7['rango edad']=='  65 years and over'].groupby(['fecha'])[['total personas']].sum().reset_index()

desem_rangosa = pd.merge(desem_rango1a, desem_rango2a, how = 'left', on = 'fecha').merge(desem_rango3a, how = 'left', on = 'fecha').merge(desem_rango4a, how = 'left', on = 'fecha').merge(desem_rango5a, how = 'left', on = 'fecha').merge(desem_rango6a, how = 'left', on = 'fecha')
desem_rangosa.columns = ['año','15-19','20-24','25-44','45-54','55-64','65 years and over']
desem_rangosa # desempleo por año segun el rango de edad 
```

```python
# el gráfico anterior en lineas
fig = px.line(desem_rangosa, x = 'año', y =['15-19','20-24','25-44','45-54','55-64','65 years and over'],
             title= '<b>desempleabilidad por año con base en los grupos por edades<b>',
             color_discrete_sequence=px.colors.qualitative.Set1)

fig.update_layout(
    xaxis_title = 'año',
    yaxis_title = 'desempleados',
    template = 'simple_white',
    title_x = 0.5)
fig.show()
```

**Interpretacion**: las personas que estan en el grupo de edad de 65 años y mas son las que presentan menor desempleo, es posible que la razon sea porque las personas de esta edad tienen menor necesidad de conseguir trabajo,ya sea por haber llegado a la edad de jubilacion o son empleados que han alcanzado una estabilidad mayor en sus empleos. por otro lado las personas que esta en el grupo de edad de 25 a 44 años son los que mayor desempleo presentan con una diferencia significativa en comparacion con los otros grupos de edad, en este caso, la situacion puede ser contraria a los de mayor edad, las personas de este rango de edad suman a las presonas que inician su vida laboral despues de los estudios, la edad de madurez e independencia y por lo tanto mayor necesidad de ingresos y un rango de edad donde la mayoria tiene un alto grado de inexperiencia laboral.
